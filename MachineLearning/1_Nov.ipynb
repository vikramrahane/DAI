```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import Ridge, Lasso, ElasticNet
from sklearn.metrics import mean_squared_error as mse
from sklearn.model_selection import KFold, cross_val_score
```

```python
boston=pd.read_csv("Boston.csv")
X=boston.drop('medv',axis=1)
y=boston['medv']
```

```python
kfold=KFold(n_splits=5, shuffle=True, random_state=23)
elastic= ElasticNet(alpha=0.1,l1_ratio=0.3)
results= cross_val_score(elastic, X, y, cv=kfold,scoring='neg_mean_squared_error')
print(results.mean())
```

### Grid Search CV

```python
from sklearn.model_selection import GridSearchCV
```

```python
l1_rat=[0.1,0.25,0.5,0.8,0.9]
alphas= [0.1,0.5,1,12,2.5,3]
params= {'alpha': alphas,'l1_ratio':l1_rat}
elastic =ElasticNet()
gcv= GridSearchCV(elastic, param_grid=params,cv=kfold,scoring='neg_mean_squared_error')
gcv.fit(X,y)
print(gcv.best_params_)
print(gcv.best_score_)
```

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import Ridge,Lasso, ElasticNet, LogisticRegression
from sklearn.metrics import mean_squared_error,log_loss, accuracy_score
```

```python
bruput= pd.read_csv("data.csv")
X= bruput.drop('Bankrupt?',axis=1)
y= bruput['Bankrupt?']
```

```python
print(y.value_counts(normalize=True)*100)

X_train, X_test, y_train, y_test=train_test_split(X,y,test_size=0.3,random_state=23,stratify=y)
```

```python
print(y_train.value_counts(normalize=True)*100)
print(y_test.value_counts(normalize=True)*100)
```

```python
lr = LogisticRegression()
lr.fit(X_train,y_train)
y_pred_prob= lr.predict_proba(X_test)[:,1]
print(log_loss(y_test,y_pred_prob))
```

```python
y_pred=lr.predict(X_test)
```

```python
print("The Error using logloss function", log_loss(y_test,y_pred_prob))
```

```python
print("The accuracy of the logistics Regression Model is ",accuracy_score(y_test,y_pred))
```

```python
from sklearn.model_selection import StratifiedKFold
from sklearn.model_selection import cross_val_score
```

```python
kfold= StratifiedKFold(n_splits=5,shuffle=True,random_state=23)
lr= LogisticRegression()
```

```python
#Default scoring-acc score
results= cross_val_score(lr,X,y,cv=kfold)
print(results.mean())
```

```python
#log loss
results= cross_val_score(lr,X,y,cv=kfold,scoring='neg_log_loss')

print(results.mean())
```

```python
# Grid Search CV

lr=LogisticRegression()
params={'penalty':['l1','l2','elasticnet',None],'solver':['lbfgs','liblinear','newton-cg','newton-cholesky','sag','saga']}
#Default scoring-accscore
gcv= GridSearchCV(lr,param_grid=params,cv=kfold)
gcv.fit(X,y)
print(gcv.best_params_)
print(gcv.best_score_)
```

```python
#log loss
gcv =GridSearchCV(lr,param_grid=params,cv=kfold,scoring='neg_log_loss')
gcv.fit(X,y)
print(gcv.best_params_)
print(gcv.best_score_)
```
