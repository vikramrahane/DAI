```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix, classification_report, accuracy_score
from sklearn.naive_bayes import BernoulliNB
from sklearn.model_selection import StratifiedGroupKFold,GridSearchCV, StratifiedKFold
```

```python
cancer=pd.read_csv('Cancer.csv')
dum_cancer= pd.get_dummies(cancer,drop_first=True)
dum_cancer.head()
```

```python
X = dum_cancer.drop('Class_recurrence-events', axis=1)
y= dum_cancer['Class_recurrence-events']
```

```python
nb= BernoulliNB()
kfold= StratifiedKFold(n_splits=5, shuffle=True, random_state=23)
params = {'alpha': np.linspace(0,5,20)}
gcv= GridSearchCV(nb,param_grid=params, cv=kfold, scoring='neg_log_loss')
```

```python
gcv.fit(X,y) # Model Building : Apriori Probs Calculation
```

```python
print(gcv.best_params_)
```

```python
print(gcv.best_score_)
```
