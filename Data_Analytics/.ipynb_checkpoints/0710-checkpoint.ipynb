```python
# Create Function to display distribution pairplot
def distribution(dataset,variable):
    '''
        Args:
            dataset: Include the dataframe here
            variable: Include the column from dataframe used for color encoding
            Returns: sns pairplot with color encoding
        
        pairplot is a function in seaborn that creates a grid of scatterplots, showing pairwise 
    '''
    g=sns.pairplot(data=dataset,hue=variable)
    # g,fig is used to access the figre (the entire plotting area) created by pairplot function
    #.suptitle is a method to set the title of the figure
    g.fig.suptitle('Graph showing distribution between score and {}'.format(variable),fontsize=20)
    # .subplots_adjust is a method that adjust
    g.fig.subplots_adjust(top=0.9)
    return g
```

```python
'''
    Finding correlation between categorical variables and test scores using 1-way ANOVA
    1-way Hypothesis:
    Null Hypothesis(H0): There is no difference between groups and equality between means
    Alternative Hypothesis(H1): There is a difference between means and groups
'''
```

```python
# Clean up column names for statsModels
df.columns=['gender','race','parental_edu','lunch','test_prep_course','math_score','reading_score','writing_score']
```

```python
# Create anova Test Function
def anova_test(data, variable):
    '''
        Args:
            dataset: (Dataframe)
            variable: Categorical column used for 1-way ANOVA test
            Returns: Nothing
    '''
    x=['math_score','reading_score','writing_score']
    for i,k in enumerate(x):
        lm=ols('{} ~ {}'.format(x[i],variable),data=data).fit()
        table=sm.stats.anova_lm(lm)
        print("P-value for 1-way ANOVA Test between {} and {} is".format(x[i],variable),table.loc[variable,"PR(>F)"])
```

 # Skewness

```python
from scipy import stats
```

```python
# Generate  normal distribution with 10000 data points, plot it , show statistics
np.random.seed(33)
norm_sample=np.random.randn(10000)
sns.histplot(x=norm_sample,kde=True);
plt.show()
```

```python
print("mean=",np.mean(norm_sample))
print("median=",np.median(norm_sample))
print("std=",np.std(norm_sample))
```

```python
# Generate a sample of 10000 data points from a right-skewed distribution
# First parameter is 0 means no skewness, +ve means positive Skewness, -ve means negative Skewness
# This means the generated data will have +ve skew, i.e. long tail on the right side, and the majority of the data
# points will be concentrated on the left side

# 10 here is just 10 units, if you increase it, the right skewness will increase even further
skewed_data_r=stats.skewnorm.rvs(10,size=10000)
sns.histplot(x=skewed_data_r,kde=True);
plt.show()
```

```python
print("mean=",np.mean(skewed_data_r))
print("median=",np.median(skewed_data_r))
```

```python
# 100 here is just 100 units, if you increase it, the right skewness will increase even further
skewed_data_r=stats.skewnorm.rvs(100,size=10000)
sns.histplot(x=skewed_data_r,kde=True);
plt.show()
```

```python
print("mean=",np.mean(skewed_data_r))
print("median=",np.median(skewed_data_r))
```

```python
# Generate a sample of 10000 data points from a left-skewed distribution
# loc (location parameter): 5, which represents the mean of the distribution. This parameter sets the center of
# locationof the distribution

# In the right skewed, mean is to the right of the peak value
skewed_data_l=stats.skewnorm.rvs(-10,size=10000,loc=5)
sns.histplot(x=skewed_data_l,kde=True);
plt.show()
```

```python
print("mean=",np.mean(skewed_data_l))
print("median=",np.median(skewed_data_l))
```

```python
# Q-Q plots
stats.probplot(norm_sample,plot=plt)
plt.show()
```

```python
stats.probplot(skewed_data_r,plot=plt)
plt.show()
```

```python
stats.probplot(skewed_data_l,plot=plt)
plt.show()
```

```python
# print the skewness of each datasaet
print(stats.skew(norm_sample))
print(stats.skew(skewed_data_r))
print(stats.skew(skewed_data_l))
```

```python
# perform a skewness test on each dataset
# Returns a numerical value representing the skewness of the dataset
print(stats.skewtest(norm_sample))
print(stats.skewtest(skewed_data_r))
print(stats.skewtest(skewed_data_l))
```

```python
''' Output:
        Statistics: The statistic value is the test statistic calculated by the skewness test. It quantifies 
        the degree of skewness in the dataset.
        If negative, it indicates that the dataset is left-skewed or negatively skewed
        
        The p-value measures the evidence against a null hypothesis.
        A small p-value (typically less than 0.05) suggest that there is strong evidence to reject the null 
        hypothesis, indicating that the data is significantly skewed.
        
        A Large p-value (greater than 0.05) suggest that there is not enough evidance to reject the null 
        hypothesis, indicating that the data may not be significantly skewed
''' 
```

# Chi-Square Test

```python
'''
    Here we use stats.chisquare
    Primarilly used for conducting a chi-square goodness-of-fit test, which compares observed frequencies with 
    expected frequencies under a specified theoritcal distribution
'''
```

```python
# Fake demographic data for US and Minnesota state and walk through the chi-square goodness of fit test to 
# check whether they are different:
national=pd.DataFrame(["white"]*100000+["hispanic"]*60000+["black"]*50000+["asian"]*15000+["other"]*35000)
minnesota=pd.DataFrame(["white"]*600+["hispanic"]*300+["black"]*250+["asian"]*75+["other"]*150)
```

```python
print(national.sample(5))
print(minnesota.sample(5))
```

```python
# Create Frequence Tables (crosstabs) for both datasets using pd.crostabs. 
# national_table and minnesota_table store the counts of each demographic category
national_table=pd.crosstab(index=national[0],columns='count')
minnesota_table=pd.crosstab(index=minnesota[0],columns='count')
```

```python
print("NAtional")
print(national_table)
print(" ")
print("Minnesota")
print(minnesota_table)
```

```python
observed=minnesota_table
national_ratios=national_table/len(national) # Get population ratios
national_ratios
```

```python
# calculate the expected counts by multiplying the population ratios in the national dataset by the total no of 
# observations in the Minnesota dataset.
expected=national_ratios*len(minnesota) # Get expected counts
```

```python
# calculate the chi-squared statistic by comparing the observed and expectd counts and summing the squared diferences
# normalized by the expected counts.
chi_squared_stat= (((observed-expected)**2)/expected).sum()
```

```python
print("Calculated observed value")
print(chi_squared_stat)
```

```python
# Find the critical value for a 95% confidence level with 4 degree of freedom
crit=stats.chi2.ppf(q=0.95,df=4) # df=no of variable categories - 1 
```

```python
print("Critical Value: ",crit)
```

```python
# calculate p-value based on the chi-squared statistic and degrees of freedom
p_value=1-stats.chi2.cdf(x=chi_squared_stat, df=4)
print("P-value: ",p_value)
```

```python
'''
    Since our chi-squared statistic exceeds the critical value, we'd reject the null hypothesis that the two 
    distribution are the same.
    You can carry out a chi-squared goodness-of-fit test automatically using the scipy function scipy.stats.chisquare()
'''
print("Observed value through stats library")
print(stats.chisquare(f_obs=observed,f_exp=expected))  # Array of observed counts
```

```python
# The test result agree with the values we calculated above.
```

# ch2_contingency

```python
'''HEre we use ch2_contingency
    Purpose: It is used for conducting a ch-square test of independence or association between two categorical 
    variables by analyzing a contigency table (also known as a two-way table)
'''
```

```python
import warnings as ws
from scipy.stats import chi2_contingency
```

```python
df=pd.read_csv("heart_failure_clinical_records_dataset.csv")
df.head()
```

```python
print("Against Age")
print(chi2_contingency(pd.crosstab(index=df['age'],columns=df['DEATH_EVENT'])))
```

```python
print("Against Anemia")
print(chi2_contingency(pd.crosstab(index=df['anaemia'],columns=df['DEATH_EVENT'])))
```

```python
print("Against Diabetes")
print(chi2_contingency(pd.crosstab(index=df['diabetes'],columns=df['DEATH_EVENT'])))
```

```python
print("Against High Blood Pressure")
print(chi2_contingency(pd.crosstab(index=df['high_blood_pressure'],columns=df['DEATH_EVENT'])))
```

```python
print("Against Sex")
print(chi2_contingency(pd.crosstab(index=df['sex'],columns=df['DEATH_EVENT'])))
```

```python
# The only column categorical that has relationship with Death Events is age, that we can also see in graph below
plt.figure(figsize=(20,6))
title=plt.title('Survival and Deaths by Age',fontsize=20)
title.set_position([0.5,1.15])
ax=sns.countplot(x="age",hue="DEATH_EVENT",data=df)
ax.set_xlabel('Age')
ax.set_ylabel('Count')
a=ax.set_xticklabels(ax.get_xticklabels(),rotation=0,horizontalalignment="center")
plt.show()
```
