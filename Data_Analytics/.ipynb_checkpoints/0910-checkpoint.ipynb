```python
from sklearn import tree
```

```python
input_file="PastHires.csv"
df=pd.read_csv(input_file,header=0)
df.head()
```

```python
# Scikit-Learn needs everything to be numerical for decision trees to work, so we need to map our data to numerical
# from using the map function
d={'Y':1,'N':0}
df['Hired']=df['Hired'].map(d)
df['Employed?']=df['Employed?'].map(d)
df['Top-tier school']=df['Top-tier school'].map(d)
df['Interned']=df['Interned'].map(d)

d={'MS':1,'BS':0,'PhD':2}
df['Level of Education']=df['Level of Education'].map(d)
df.head()
```

```python
# SeperateFeatures from the target column that we are building a tree for 
features=list(df.columns[:6])
features
```

```python
# construct the decision tree
y=df['Hired']
x=df[features]
clf=tree.DecisionTreeClassifier()
clf=clf.fit(x,y)
```

# Monte Carlo

```python
# Part 1
def roll_dice():
    # Simulation of rolling a dice twice, minimum no we get is 1, max is 6
    # Add the results of the two simulations, so possible values will be:
    # (1,1) or (1,2) ... (6,6) ....  sum will be between 2 and 12
    # Run multiple times to verify
    return np.sum(np.random.randint(1,7,2))
```

```python
print(roll_dice())
```

```python
# Part 2
```

```python
# Someonr approaches us saying I will give you 5 dollars if you get 7 and take 1 dollar if you get a no other than 7
# How do we know what will happen?
# Our own "Monte carlo simulation" like function
def monte_carlo_simulation(runs=1000):
    results=np.zeros(2,dtype=int) # An Array results[0] and results[1] initialised to two zeroes
    for _ in range(runs): # Run 1000 Times
        if roll_dice()==7:
            results[0]+=1
        else:
            results[1]+=1
    return results
```

```python
# Test 2-3 times and calculate how much you will win versus lose
print(monte_carlo_simulation())
print(monte_carlo_simulation())
print(monte_carlo_simulation())
```

```python
results=np.zeros(1000)
for i in range(1000):
    results[i]=monte_carlo_simulation()[0]
```

```python
print(results)
```

```python
np.max(results)
```

```python
# let us plot it
fig,ax=plt.subplots()
ax.hist(results,bins=15)
plt.show()
```

```python
# Our win loss
print(results.mean())          # General mean
print(results.mean()*5)        # What we will get as win on an average
print(1000-results.mean())     # What we will pay on an average
print(results.mean()/1000)     # Probability of the 'we will pay' result 
# The last probability should be close to the theoretical probability of getting a 7 when we throw two dice
```

# Example 2

```python
rev_m=170
rev_stdev=20
iterations=1000

# create a normal distribution using these
revenue=np.random.normal(rev_m,rev_stdev,iterations)
print(revenue)
```

```python
# Plot it 
plt.figure(figsize=(15,6))
plt.title('Revenue Simulations')
plt.plot(revenue)
plt.show()
```

```python
# Now costs of goods sold
# Generally it is about 60% of the revenue with 10% SD
# We already have 1000 revenues records, use them to randomly assign COGS in the above range
cogs=(revenue*np.random.normal(0.6,0.1))
```

```python
plt.figure(figsize=(15,6))
plt.title('Cost of Goods Sold Simulation')
plt.plot(cogs)
plt.show()
```

```python
# calculate grod Profit
gross_profit=revenue-cogs
print(gross_profit)
```

```python
plt.figure(figsize=(15,6))
plt.title('Gross profit Simulation')
plt.plot(gross_profit)
plt.show()
```

```python
# create a stacked bar chart
numbers=list(range(1,1001))
```

```python
plt.figure(figsize=(10,6))
plt.bar(numbers,revenue,label='Revenue',color='skyblue')
plt.bar(numbers,cogs,label='cogs',color='lightcoral')
plt.bar(numbers,gross_profit,bottom=[r+c for r,c in zip(revenue,cogs)],label='Gross Profit',color='lightgreen')
plt.xlabel('Number of Times')
plt.ylabel('Amount($)')
plt.title('Revenue, COGS and Gross Profit using Monte Carlo')
plt.legend()
plt.show()
```

# PCA

```python
from sklearn.model_selection import cross_val_score
from sklearn.preprocessing import LabelEncoder
from sklearn.decomposition import PCA
from sklearn import metrics
from sklearn import linear_model
from sklearn.model_selection import train_test_split
import time
```

```python
# Students performance in MAths
student_data=pd.read_csv("student-mat.csv")
```

```python
student_data.describe()
```

```python
col_str=student_data.columns[student_data.dtypes==object]
print(col_str)
```

```python
# Convert each category value into a new column assign a 1 or 0 (True/False) value to the column. This has the benefit
# of not weighting a value improperly. Simplest method is using pandas ' .get_dummies()' method
# drop-first=true , reduces extra column creation 
student_data=pd.get_dummies(student_data,columns=col_str,drop_first=True)
print(student_data.info())
```

```python
print(student_data[['G1','G2','G3']].corr())
# since G1,G2,G3 have very high correlation we can drop G1,G2,G3
```

```python
student_data.drop(axis=1,labels=['G1','G2'])
```

```python
# Drop the G3 column, beccause we want to predict it now
label=student_data['G3'].values
predictors=student_data.drop(axis=1,labels=['G3']).values
student_data.shape
```

```python
# Using Linear Regression to Predict Grades
lr=linear_model.LinearRegression()
```

```python
'''
    cross_val_score: Used during the testing and validation phase of your machine learning model development
    Trains and tests a model over multiple folds of your dataset. This cross validation method gives you a 
    better understanding of model performance over the whole dataset instead of just a single train/test split.
    
    The no of folds is defined, by default this is 5
    The dataset is split up according to these folds, where each fold has a unique set of testing data
    A model is trained and tested for each fold
    Each fold returns a metric for it's test data
    The mean and standard deviation of these metrics can then be calculated to provide a single metric for the process
'''

#Returns an array of scores of the estimator for each run of the cross validation
lr_score=cross_val_score(lr,predictors,label,cv=5) # 5 runs, 5 means
print("LR model cross validation score: "+str(lr_score))
# Average of all these means
print("LR model cross validation mean score: "+str(lr_score.mean()))
```

```python
# Using PCA
from sklearn.decomposition import PCA
pca=PCA(n_components=len(student_data.columns)-1)
pca.fit(predictors)
variance_ratio=pca.explained_variance_ratio_
print(pca.explained_variance_.shape)
```

```python
# Now plot
# Find cumulative variance, adding one independent variable at a time
variance_ratio_cum_sum=np.cumsum(variance_ratio)
print(variance_ratio_cum_sum)
```

```python
# This cumulative explained variance graph helps us to choose the no of desired principle components. If we look at 
# the above print statement's output, we will realize that 90% variations in the data is explaining by the first 
# 6 principal components. Hence, we annotate 6 on the graph

plt.plot(variance_ratio_cum_sum)
plt.xlabel('No of Components')
plt.ylabel('cumulative explained variance')

# Annotate 90% variance explaining by the first 6 variables only
plt.annotate('6',xy=(6,.90))
plt.show()
```

```python
# individual explained variance, instead of cummulative variance
# We see that the first variable causes 60% variance, the 2nd 22%, and so on ...
plt.figure(figsize=(10,5))
plt.bar(range(41),pca.explained_variance_,alpha=0.5,label='individual explained variance')
plt.xlabel('explained variance ratio')
plt.ylabel('Principal Components')
plt.legend(loc='best')
plt.show()
```

```python
# PCA transform a set of correlated variables into a set of linearly uncorelated variables called principal 
# components, we can check the correlation with a heat map of correlation matrix

# Correlation between the variables of th original data
# we see high correlation, which mean duplication - we can eliminate many variables
coorelation=pd.DataFrame(predictors).corr()
sns.heatmap(coorelation,vmax=1,square=True,cmap='Greens')
plt.title('Correlation between different features')
plt.show()
```

```python
# Looking at the above plot we are taking 6 variables
pca=PCA(n_components=6)
pca.fit(predictors)
Transformed_vector=pca.fit_transform(predictors)
print(Transformed_vector)
```

```python
# correlation between 6 variables after transforming the dat with PCA is 0
coorelation=pd.DataFrame(Transformed_vector).corr()
sns.heatmap(coorelation,vmax=1,square=True,cmap='viridis')
plt.title('Correlation between different features')
plt.show()
```

```python
# Check the performance with the 6 variables
lr_pca=linear_model.LinearRegression()
lr_pca_score=cross_val_score(lr_pca,Transformed_vector,label,cv=5)
print("PCA Model Cross Validation Score "+str(lr_pca_score))
print("PCA Model Cross Validation Mean Score "+str(lr_pca_score.mean()))
```

```python
# We see the values similar to the earlier case when we had 40 independent variables.
# This means that pca has indeed reduced 40 variables to 6 without causing any negative impact
```

# ARIMA

```python
from statsmodels.tsa.arima.model import ARIMA
```

```python
# sample_dataset 
'''
    Data dictionary defined a time series dataset with three components( contains 3 key values pairs)
    Month :  the time index or time span 
    Sales : The variable we want to forecast 
    AdvertisingSpend : An external variable that may impact the 'sale' variable, used in SARIMAX model
'''
data={'Month':pd.date_range(start='2019-01-01',periods=36,freq='M'),
      'Sales':[100,120,130,150,200,220,250,230,210,180,160,140,110,130,150,190,220,240,260,250,230,200,180,160,130,140,
              160,190,210,240,270,250,230,200,210,190],
     'AdvertisingSpend':[50,60,70,80,90,100,110,120,130,140,150,160,60,70,80,90,100,110,120,130,140,150,160,
                        70,80,90,100,110,120,130,140,150,160,70,120,90]}
```

```python
df=pd.DataFrame(data=data)
df.set_index('Month',inplace=True)
```

```python
# Fit an ARIMA model to the sales data
order=(2,1,1) # (p,d,q) order
model=ARIMA(df['Sales'],order=order)
results=model.fit()
```

```python
# Forecast features sales values
forecast_steps=12
forecast=results.forecast(steps=forecast_steps)
```

```python
# Plot the original sales data and ARIMA forcast
plt.figure(figsize=(12,6))
plt.plot(df['Sales'],label='Actual Sales')
plt.plot(pd.date_range(start=df.index[-1],periods=forecast_steps,freq='M'),forecast,label='ARIMA forecast',color='red')
plt.xlabel('Month')
plt.ylabel('Sales')
plt.title('ARIMA Forecast for sales')
plt.legend()
plt.show()
```

```python
# Fit a SARIMAX model to the sales from data AdvertisingSpend as an exogenous variable 
from statsmodels.tsa.statespace.sarimax import SARIMAX
```

```python
'''
    SARIMAX allows incorporate additional external factors (exogeneous variable) that can be influence the time series
    we are trying to forecast in this case we are using the 'AdvertisingSpend' variable as an exogeneous variable
    if this variable has significant impact on sales, including it in the model can improve the accuracy of 
    the forecast.
    
    SARIMAX allows us to specify both non-seasonal (p,d,q) and seasonal (P,D,Q,s) orders. This flexibility enables us
    to tailor the model to the specific characteristic of our data. In the povided code we have chosen an order of 
    (2,1,1) and seasonal_order of (1,1,1,12) which may better capture data's behaviour compare to the ARIMA order.
'''
order=(2,1,1) # (p,d,q)
seasonal_order=(1,1,1,12) # {P,D,Q,s}
```

```python
exog=df['AdvertisingSpend']
model=SARIMAX(df['Sales'],order=order,exog=exog,seasonal_order=seasonal_order)
results=model.fit()
```

```python
# Forecast feature 'Sales' values
forecast_steps=24
forecast=results.get_forecast(steps=forecast_steps,exog=df.iloc[-forecast_steps:]['AdvertisingSpend'])
```

```python
# Plot the original sales data, AdvertisingSpend and SARIMAX
plt.figure(figsize=(12,6))
plt.plot(df['Sales'],label='Actual Sales')
plt.plot(pd.date_range(start=df.index[-1],periods=forecast_steps,freq='M'),forecast.predicted_mean,label='SARIMAX forecast',color='green')
plt.plot(df.index[-forecast_steps:],df['AdvertisingSpend'][-forecast_steps:],label='AdvertisingSpend',linestyle='--',color='red')
plt.xlabel('Month')
plt.ylabel('Sales/AdvertisingSpend')
plt.title('SARIMAX Forecast for sales')
plt.legend()
plt.show()
```
