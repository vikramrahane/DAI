<h3>Creating grand alliance</h3>

```python
NDA = ['BJP','SS', 'LJP', 'SAD', 'RLSP','AD','PMK','NPP','AINRC','NPF','RPI(A)','BPF','JD(U)','SDF',
       'NDPP','MNF','RIDALOS','KMDK','IJK','PNK','JSP','GJM','MGP','GFP','GVP','AJSU','IPFT','MPP',
       'KPP','JKPC','KC(T)','BDJS','AGP','JSS','PPA','UDP','HSPDP','PSP','JRS','KVC','PNP','SBSP',
       'KC(N)','PDF','MDPF']

UPA = ['INC','NCP', 'RJD', 'DMK', 'IUML','JMM','JD(s)','KC(M)','RLD','RSP','CMP(J)','KC(J)','PPI','MD']

Others = ['YSRCP','AAAP', 'IND', 'AIUDF', 'BLSP', 'JKPDP', 'JD(S)', 'INLD', 'CPI', 'AIMIM', 'KEC(M)',
          'SWP', 'NPEP', 'JKN', 'AIFB', 'MUL', 'AUDF', 'BOPF', 'BVA', 'HJCBL', 'JVM','MDMK']
```

```python
candidates_09['Alliance'] = candidates_09['Party Abbreviation'].map(lambda x: 'NDA' if x in NDA 
                                                                    else ('UPA' if x in UPA 
                                                                          else 'Others'))
```

<h3>Gender Distributions</h3>

```python
def plot_gender_distribution(df, y):
    gender_values = [df['Candidate Sex'].value_counts()[0], 
                     df['Candidate Sex'].value_counts()[1]]
    gender_label = ['M', 'F']

    plt.figure(figsize=(9, 6))
    plt.pie(gender_values, labels=gender_label, frame = True, shadow=True, colors = ['skyblue', 'pink'],
            autopct= lambda x: '{:.0f}'.format(x*np.sum(gender_values)/100));
```

```python
candidates_09['Candidate Sex'].value_counts()
```

```python
plot_gender_distribution(candidates_09, '09')
plt.title("Gender Distributions of Candidates in 2009")
```

```python
plot_gender_distribution(candidates_14, '14')
plt.title("Gender Distributions of Candidates in 2014");
```

```python
plot_gender_distribution(nda_09, '09')
plt.title("Gender distribution of NDA candidates in 2009");
```

```python
plot_gender_distribution(upa_09, '09')
plt.title("Gender distribution of UPA candidates in 2009");
```

```python
plot_gender_distribution(nda_14, 14)
plt.title("Gender distribution of NDA candidates in 2014");
```

```python
plot_gender_distribution(upa_14, 14)
plt.title("Gender distribution of UPA candidates in 2014");
```
