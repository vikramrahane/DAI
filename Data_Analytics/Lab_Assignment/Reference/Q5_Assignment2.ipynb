# Q2. Create Winning seats distribution by Major Political Parties & Alliances for 2009 & 2014 Q3. How many seats won by Alliances and Major Political Parties ?

# Q5. Plot the age distribution of winners of both 2014 and 2009 elections

```python
#Plotting age distribution for (2009 election and 2014)
sns.distplot(cd09_win_Major['Candidate Age'],label='Age 2009',color='r',hist=None);
sns.distplot(cd14_win_Major['Candidate Age'],label='Age 2014',color='b',hist=None);
plt.legend();
```

# Q6. Separately plot age distribution of NDA & UPA candidates

```python
# Age distrobution plot (2009 election)
cd09_NDA=cd09_win_Major[cd09_win_Major['Alliance']=='NDA']
cd09_UPA=cd09_win_Major[cd09_win_Major['Alliance']=='UPA']
sns.kdeplot(cd09_NDA['Candidate Age'],label='NDA 2009',fill=True)
sns.kdeplot(cd09_UPA['Candidate Age'],label='UPA 2009',fill=True)
plt.legend();
```

```python
# Age distrobution plot (2014 election)
cd14_NDA=cd14_win_Major[cd14_win_Major['Alliance']=='NDA']
cd14_UPA=cd14_win_Major[cd14_win_Major['Alliance']=='UPA']
sns.kdeplot(cd14_NDA['Candidate Age'],label='NDA 2014',fill=True)
sns.kdeplot(cd14_UPA['Candidate Age'],label='UPA 2014',fill=True)
plt.legend();
```

# Q10. Plot the poll percentage of states for 2009 & 2014 elections

```python
cd09_state_votes=pd.DataFrame(cd09.groupby('State name')['Total Votes Polled'].sum())
cd09_state_votes['Total Votes Polled'].sum()
cd09_state_votes['Percent Votes']=(cd09_state_votes['Total Votes Polled']/(cd09_state_votes['Total Votes Polled'].sum()))*100
sns.barplot(x=cd09_state_votes.index,y=cd09_state_votes['Percent Votes']);
plt.xticks(rotation=80);
```
