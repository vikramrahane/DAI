```python
'''
    Indians and elections are the things which keeps on happening almost every month and every year.
Well, Loksabha elections is one such which happens after every 5 years, well we have data of
candidates and electors from 2009 and 2014. Compare the dataset and compute the results visually
1. Create grand alliances
a. 'INC','NCP', 'RJD', 'DMK', 'IUML',
    'JMM','JD(s)','KC(M)','RLD','RSP','CMP(J)','KC(J)','PPI','MD' as UPA
b. 'BJP','SS', 'LJP', 'SAD', 'RLSP',
    'AD','PMK','NPP','AINRC','NPF','RPI(A)','BPF','JD(U)','SDF','NDPP','MNF','RIDALOS','KMDK',
    'IJK','PNK','JSP','GJM','MGP','GFP','GVP','AJSU','IPFT','MPP','KPP','JKPC','KC(T)','BDJS',
    'AGP','JSS','PPA','UDP','HSPDP','PSP','JRS','KVC','PNP','SBSP','KC(N)','PDF','MDPF' as NDA
c. 'YSRCP','AAAP', 'IND', 'AIUDF', 'BLSP', 'JKPDP', 'JD(S)', 'INLD', 'CPI', 'AIMIM',
    'KEC(M)','SWP', 'NPEP', 'JKN', 'AIFB', 'MUL', 'AUDF', 'BOPF', 'BVA', 'HJCBL',
    'JVM','MDMK' as Others
'''
```

```python
import pandas as pd
```

```python
candidate2009=pd.read_csv("LS2009Candidate.csv")
candidate2014=pd.read_csv("LS2014Candidate.csv")
```

```python
candidate2009.head()
```

```python
candidate2014.head()
```

```python
candidate2009['Alliance']=candidate2009['Party Abbreviation']
candidate2014['Alliance']=candidate2014['Party Abbreviation']
candidate2009.head()
```

```python
candidate2014.head()
```

```python
UPA=['INC','NCP', 'RJD', 'DMK', 'IUML','JMM','JD(s)','KC(M)','RLD','RSP','CMP(J)','KC(J)','PPI','MD']
NDA=['BJP','SS', 'LJP', 'SAD', 'RLSP','AD','PMK','NPP','AINRC','NPF','RPI(A)','BPF','JD(U)','SDF','NDPP','MNF',
     'RIDALOS','KMDK','IJK','PNK','JSP','GJM','MGP','GFP','GVP','AJSU','IPFT','MPP','KPP','JKPC','KC(T)','BDJS',
     'AGP','JSS','PPA','UDP','HSPDP','PSP','JRS','KVC','PNP','SBSP','KC(N)','PDF','MDPF']
Others=['YSRCP','AAAP', 'IND', 'AIUDF', 'BLSP', 'JKPDP', 'JD(S)', 'INLD', 'CPI', 'AIMIM','KEC(M)','SWP', 'NPEP',
        'JKN', 'AIFB', 'MUL', 'AUDF', 'BOPF', 'BVA', 'HJCBL','JVM','MDMK']
```

```python
candidate2009['Alliance']=candidate2009['Alliance'].map(lambda x: 'UPA' if x in UPA else 'NDA' if x in NDA else 'Others')
```

```python
candidate2009.head(10)
```

```python
candidate2014['Alliance']=candidate2014['Alliance'].map(lambda x: 'UPA' if x in UPA else 'NDA' if x in NDA else 'Others')
candidate2014.head()
```

## 2. Create Winning seats distribution by Major Political Parties & Alliances for 2009 & 2014
## 3. How many seats won by Alliances and Major Political Parties ?

```python
winseat_dist_party09=candidate2009[candidate2009['Position']==1]
winseat_dist_party09.shape
```

```python
winseat_dist_party14=candidate2014[candidate2014['Position']==1]
winseat_dist_party14.shape
```

```python
winseat_dist_party09.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
major_win=winseat_dist_party09.groupby('Party Abbreviation')['Position'].count().sort_values(ascending=False).head(15)
major_win
```

```python
sns.barplot(x=major_win.index,y=major_win.values)
plt.ylabel("Seats")
plt.xlabel("Party")
plt.title("No of Winning Seats of Major Parties in 2009")
plt.show()
```

```python
alliance_major_win=winseat_dist_party09.groupby('Alliance')['Position'].count().sort_values(ascending=False)
alliance_major_win
```

```python
sns.barplot(x=alliance_major_win.index,y=alliance_major_win.values)
plt.ylabel("Seats")
plt.xlabel("Alliance")
plt.title("No of Winning Seats of Different Alliances in 2009")
plt.show()
```

```python
major_win_14=winseat_dist_party14.groupby('Party Abbreviation')['Position'].count().sort_values(ascending=False).head(15)
major_win_14
```

```python
sns.barplot(x=major_win_14.index,y=major_win_14.values)
plt.ylabel("Seats")
plt.xlabel("Party")
plt.title("No of Winning Seats of Major Parties in 2014")
plt.show()
```

```python
alliance_major_win_14=winseat_dist_party14.groupby('Alliance')['Position'].count().sort_values(ascending=False)
alliance_major_win_14
```

```python
sns.barplot(x=alliance_major_win_14.index,y=alliance_major_win_14.values)
plt.ylabel("Seats")
plt.xlabel("Alliance")
plt.title("No of Winning Seats of Different Alliances in 2014")
plt.show()
```

## 4. Plot comparatively seats won based on candidate category as General, ST and SC for 2009 & 2014

```python
data09=winseat_dist_party09.groupby('Candidate Category')['Position'].count()
data09
```

```python
plt.pie(data09.values,labels=data09.index,startangle=90,autopct='%0.2f',explode=(0.1,0.1,0))
plt.title("seats won based on candidate category as General, ST and SC for 2009")
plt.show()
```

```python
data14=winseat_dist_party14.groupby('Candidate Category')['Position'].count()
data14
```

```python
plt.pie(data14.values,labels=data14.index,startangle=90,autopct='%0.2f',explode=(0.1,0.1,0))
plt.title("seats won based on candidate category as General, ST and SC for 2014")
plt.show()
```

## 5. Plot the age distribution of winners of both 2014 and 2009 elections

```python
sns.displot(winseat_dist_party09['Candidate Age'],label='Age 2009')
sns.displot(winseat_dist_party14['Candidate Age'],label='Age 2014')
plt.title("age distribution of winners of 2009 & 2014 elections")
plt.legend()
plt.show()
```

```python
sns.distplot(winseat_dist_party09['Candidate Age'],hist=None,label='Age 2009')
sns.distplot(winseat_dist_party14['Candidate Age'],hist=None,label='Age 2014')
plt.title("age distribution of winners of 2009 & 2014 elections")
plt.legend()
plt.show()
```

## 6. Separately plot age distribution of NDA & UPA candidates

```python
cand09_NDA=winseat_dist_party09[winseat_dist_party09['Alliance']=='NDA']
cand09_UPA=winseat_dist_party09[winseat_dist_party09['Alliance']=='UPA']
sns.kdeplot(cand09_NDA['Candidate Age'],label='NDA 2009',fill=True)
sns.kdeplot(cand09_UPA['Candidate Age'],label='UPA 2009',fill=True)
plt.legend()
plt.plot()
```

```python
cand14_NDA=winseat_dist_party14[winseat_dist_party14['Alliance']=='NDA']
cand14_UPA=winseat_dist_party14[winseat_dist_party14['Alliance']=='UPA']
sns.kdeplot(cand14_NDA['Candidate Age'],label='NDA 2014',fill=True)
sns.kdeplot(cand14_UPA['Candidate Age'],label='UPA 2014',fill=True)
plt.legend()
plt.plot()
```

## 10. Plot the poll percentage of 'states' for 2009 & 2014 elections

```python
candidate2009.head()
```

```python
state_09=candidate2009.groupby('State name')['Total Votes Polled'].sum()
d09=state_09.map(lambda x: x/candidate2009['Total Votes Polled'].sum()*100)
sns.barplot(x=d09.index,y=d09.values)
plt.xticks(rotation=90)
plt.title("poll percentage of 'states' for 2009 ")
plt.show()
```

```python
state_14=candidate2014.groupby('State name')['Total Votes Polled'].sum()
d14=state_14.map(lambda x: x/candidate2014['Total Votes Polled'].sum()*100)
sns.barplot(x=d14.index,y=d14.values)
plt.xticks(rotation=90)
plt.title("poll percentage of 'states' for 2014 ")
plt.show()
```
