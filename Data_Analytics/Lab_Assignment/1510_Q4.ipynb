### Q1. Find out the alternative sources of fuel available in 2008 & 2018 respectively and by how much?

```python
import matplotlib.pyplot as plt
```

```python
fuel08=clean_08.groupby('fuel')['displ']
fuel_cnt=fuel08.count()
fuel_cnt
```

```python
fuel_total=fuel08.count().sum()
fuel_total
```

```python
fuel_per=fuel_cnt.map(lambda x:x/fuel_total*100)
fuel_per
```

```python
plt.bar(fuel_per.index,fuel_per.values)
plt.show()
```

```python
''' Coclusion: By observing above we can see that there are 4 types of fuel used in 2008 and Gasoline used 
    99.69 % times. So Gasoline is Most used fuel.
'''
```

```python
fuel18=clean_18.groupby('fuel')['displ']
fuel_cnt=fuel18.count()
fuel_cnt
```

```python
fuel_total=fuel18.count().sum()
fuel_total
```

```python
fuel_per=fuel_cnt.map(lambda x:x/fuel_total*100)
fuel_per
```

```python
plt.bar(fuel_per.index,fuel_per.values)
plt.show()
```

```python
''' Coclusion: By observing above we can see that there are 5 types of fuel used in 2018 and Gasoline used 
    90.02 % times. So Gasoline is Most used fuel. 
    There is significant change in percent use of other fuel types also. 
    New 2(Diesel & Electricity) Types of fuel are come in trend. 
'''
```

## Q2. Is there any improvement in â€˜fuel economyâ€™ with respect to vehicle class from 2008 to 2018, perform the necessary steps to examine?

```python
clean_08.veh_class.unique()
```

```python
clean_18.veh_class.unique()
```

```python
veh_class_08=clean_08.groupby('veh_class')['model']
print("Count of vehicle as per class in 2008: \n",veh_class_08.count())
veh_class_18=clean_18.groupby('veh_class')['model']
print("Count of vehicle as per class in 2018: \n",veh_class_18.count())
```

```python
d={'Vehicle Class':veh_class_08.count().index,'2008':veh_class_08.count().values}
d08=pd.DataFrame(data=d)
d08
```

```python
d={'Vehicle Class':veh_class_18.count().index,'2018':veh_class_18.count().values}
d18=pd.DataFrame(data=d)
d18
```

```python
cnt_veh_class=pd.merge(d08,d18,on='Vehicle Class',how='outer')
cnt_veh_class
```

```python
cnt_veh_class.fillna(0)
```

```python
plt.figure(figsize=(12,6))
plt.bar(cnt_veh_class['Vehicle Class'],cnt_veh_class['2008'],color='green',width=0.5,label='2008')
plt.bar(cnt_veh_class['Vehicle Class'],cnt_veh_class['2018'],color='red',align='edge',width=0.5,label='2018')
plt.xticks(rotation=45)
plt.legend()
plt.title("Count of Vehicle as per Vehicle class")
plt.plot()
```

## Q5. How much improvement is there in miles/gallon or mpg? Also, which vehicle has improved the most?

```python
model_08_mpg=clean_08.groupby('model')[['city_mpg','hwy_mpg','cmb_mpg']].mean()
model_08_mpg
```

```python
d={'model':model_08_mpg.index,'city_mpg_08':model_08_mpg.city_mpg.values,
   'hwy_mpg_08':model_08_mpg.hwy_mpg.values,
   'cmb_mpg_08':model_08_mpg.cmb_mpg.values}
d08=pd.DataFrame(data=d,index=range(model_08_mpg.shape[0]))
d08
```

```python
model_18_mpg=clean_18.groupby('model')[['city_mpg','hwy_mpg','cmb_mpg']].mean()
model_18_mpg
```

```python
d={'model':model_18_mpg.index,'city_mpg_18':model_18_mpg.city_mpg.values,'hwy_mpg_18':model_18_mpg.hwy_mpg.values,
   'cmb_mpg_18':model_18_mpg.cmb_mpg.values}
d18=pd.DataFrame(data=d)
d18
```

```python
model_comb_mpg=pd.merge(d08,d18,on='model',how='outer')
model_comb_mpg
```

```python
model_comb_mpg.isna().sum()
```

```python
model_comb_mpg=model_comb_mpg.fillna(0)
model_comb_mpg
```

```python
model_comb_mpg['diff_cmb_mpg']=model_comb_mpg['cmb_mpg_18']-model_comb_mpg['cmb_mpg_08']
model_comb_mpg['diff_city_mpg']=model_comb_mpg['city_mpg_18']-model_comb_mpg['city_mpg_08']
model_comb_mpg
```

```python
model_comb_mpg[['cmb_mpg_08','cmb_mpg_18','diff_city_mpg']].describe()
```

```python
mask1=model_comb_mpg.cmb_mpg_08!=0
mask2=model_comb_mpg.cmb_mpg_18!=0

non_zero=model_comb_mpg[mask1 & mask2]
non_zero[non_zero.diff_cmb_mpg==non_zero.diff_cmb_mpg.max()]
```
