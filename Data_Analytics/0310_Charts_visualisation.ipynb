```python
plt.hist(df['Income'],edgecolor='black')
plt.show()
```

```python
plt.hist(df['Income'],bins=15,edgecolor='black')
plt.show()
```

```python
df=pd.read_csv("ex5.csv",index_col='branch number')
df
```

```python
df.plot.pie(y='sale1')
plt.show()
```

```python
#change the size to reduce clutter
df.plot.pie(y='sale1',figsize=(11,6))
plt.show()
```

```python
#Show percentage for each category
df.plot.pie(y='sale1',figsize=(11,6),autopct="%1.1f%%")
plt.show()
```

```python
#plot each column separately
df.plot.pie(subplots=True,figsize=(20,10))
plt.show()
```
