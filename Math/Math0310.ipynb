```python
import numpy as np 


xi = np.array([4,6,7,8,1,9])
yi = np.array([17,23,26,29,8,32])

eta = 0.05
b0 = 0.4 
w0 = -0.6 

ycap = xi*w0 + b0
for i in range(1,1001):

    db = -(1/len(xi))*np.sum(yi-ycap)
    dw = -(1/len(xi))*np.sum((yi-ycap)*xi)
    
    new_b = b0 - eta*db 
    new_w = w0 - eta*dw
    
    b0 = new_b 
    w0 = new_w
    ycap = xi*w0 + b0
    print("L =", (1/(2*len(xi))*np.sum((yi-ycap)**2)))
    print("w =", w0, "; b =", b0)
```

```python
insure=pd.read_csv("Insure_auto.csv")
```

```python
x1=insure['Home']
x2=insure['Automobile']
y=insure['Operating_Cost']
```

```python
x1_scaled=(x1-np.min(x1))/(np.max(x1)-np.min(x1))
x2_scaled=(x2-np.min(x2))/(np.max(x2)-np.min(x2))
y_scaled=(y-np.min(y))/(np.max(y)-np.min(y))

eta=0.4
b0=0.4
w1=-0.6
w2=0.5
nobs=insure.shape[0]
ycap=x1_scaled*w1+x2_scaled*w2+b0
for i in range(1,1001):
    db=-(1/nobs)*np.sum(y_scaled-ycap)
    dw1=-(1/nobs)*np.sum((y_scaled-ycap)*x1_scaled)
    dw2=-(1/nobs)*np.sum((y_scaled-ycap)*x2_scaled)
    
    new_b = b0 - eta*db 
    new_w1 = w1 - eta*dw1
    new_w2 = w2 - eta*dw2
    
    b0 = new_b 
    w1 = new_w1
    w2 = new_w2
    
    ycap=x1_scaled*w1+x2_scaled*w2+b0
    print("L =", (1/(2*nobs))*np.sum((y_scaled-ycap)**2))
    print("w1 =", w1,";w2= ",w2, "; b =", b0)
```
